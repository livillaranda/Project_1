```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
```

```python
# Locate CSV files
crime_2016 = "2016_crime_LIV.csv"
#crime_2018 = "crime_2018.csv"
```

```python
# Read Files
crime16 = pd.read_csv(crime_2016)

#crime18 = pd.read_csv(crime_2018)
```

```python
# View crime_2016 file
df16 = pd.DataFrame(crime16)
df16
```

```python
# Rename Columns
df16.columns = ["State", "City", "Year", "Population", "Violent Crime", "Murder", "Rape-Revised", "Rape-Legacy", "Robbery", "Assault", "Property",
              "Burglary", "Larceny", "MV Theft", "Arson"]
```

```python
df16
```

```python
#Drop commas from all values
df16 = df16.replace(',','', regex=True)
```

```python
# Specify Columns to Change
# Change Columns to float
cols = ["Population", "Violent Crime", "Rape-Revised", "Robbery", "Assault", "Property", "Burglary", "Larceny", "MV Theft"]
df16[cols] = df16[cols].astype(int)
```

```python
df16
```

```python
# Create CSV
df16.to_csv("Crime_2016_Clean.csv")
```
