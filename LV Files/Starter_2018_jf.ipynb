```python
# Locate CSV files
# crime_2016 = "2018_crime.csv"
crime_2018_cleaned = "../crime_2018_cleaned.csv"
```

```python
# Read Files
crime_2018_cleaned = pd.read_csv(crime_2018_cleaned)

#crime18 = pd.read_csv(crime_2018)
```

```python
# View crime_2016 file
df18 = pd.DataFrame(crime_2018_cleaned)
df18.head()
```

```python
# Rename Columns
df18.columns = ["State", "City", "Year", "Population", "Violent Crime", "Murder", "Rape", "Robbery", "Assault", "Property",
              "Burglary", "Larceny", "MV Theft", "Arson"]
```

```python
#Drop commas from all values
df18 = df18.replace(',','', regex=True)
```

```python
# Specify Columns to Change
# Change Columns to float

cols = ["Population", "Violent Crime", "Assault", "Property", "Burglary", "Larceny", "Arson"]
df18[cols] = df18[cols].astype(int)

```
