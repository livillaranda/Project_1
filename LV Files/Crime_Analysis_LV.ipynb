### Split orginal DataFrame to create 2 separate dfs

```python
# Use the CSV downloaded df to start basis of spliting df
state_city = crime16_df.copy()
```

```python
# Group by year, view state by state
shuffle = state_city.sort_values(by=["Year", "State"])
shuffle
```

```python
# Split df by year into 2 new dfs
split_15 = shuffle[shuffle["Year"] <= 2015]

split_16 = shuffle[shuffle["Year"] >= 2016]

# Reset index on both dfs; delete columns with old index numbers
split_15 = split_15.reset_index()
del split_15["index"]

split_16 = split_16.reset_index()
del split_16["index"]
```

```python
# Export dfs to ensure split is correct
split_15.to_csv("2015_Split_LV.csv")
split_16.to_csv("2016_Split_LV.csv")
```
