```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
```

```python
# Path for CSV file
dup_city = "Cities_Need_Cleaning.csv"
```

```python
# Read Files
city_dups = pd.read_csv(dup_city)
```

```python
# Data into df; delete extra index column
city_dup = pd.DataFrame(city_dups)
del city_dup["Unnamed: 0"]
```

```python
# View df
city_dup
```

```python
# Group by Year, View State by State; copy df to new variable
state_city = city_dup.copy()
shuffle = state_city.sort_values(by=["Year", "State"])
shuffle
```

```python
# Split df by year into 2 new dfs
split_15 = shuffle[shuffle["Year"] <= 2015]

split_16 = shuffle[shuffle["Year"] >= 2016]

# Reset index on both dfs; delete columns with old index numbers
split_15 = split_15.reset_index()
del split_15["index"]

split_16 = split_16.reset_index()
del split_16["index"]
```

```python
# Convert to CSV to look at organization
split_15.to_csv("2015_Split_Data.csv")
split_16.to_csv("2016_Splite_Data.csv")
```
