```python
# Dependencies
import pandas as pd
```

```python
# Import Data Files
crime2016 = "2016_crime_original.csv"
crime2018 = "2018_crime_original.csv"
```

```python
# Read Files
crime16 = pd.read_csv(crime2016)
crime18 = pd.read_csv(crime2018)
```

```python
# Start on 2016 Data
df16 = pd.DataFrame(crime16)
df16
```

```python
# Rename Columns
df16.columns = ["State", "City", "Year", "Population", "Violent Crime", "Murder", "Rape-Revised", "Rape-Legacy", "Robbery", "Assault",
                "Property", "Burglary", "Larceny", "MV Theft", "Arson"]
```

```python
# Remove Commas from All Values
df16 = df16.replace(',','', regex=True)
```

```python
# Specify Columns Requiring Data Type Change
cols = ["Year", "Population", "Violent Crime", "Murder", "Rape-Revised", "Rape-Legacy", "Robbery", "Assault", "Property", "Burglary",
        "Larceny", "MV Theft", "Arson"]
```

```python
# Change Specified Columns
df16[cols] = df16[cols].astype(int)
```

```python
# Merge Columns by Same Crime
mergedata = df16["Rape-Revised"] + df16["Rape-Legacy"]
df16["Rape"] = mergedata
```

```python
# Drop Original Merged Columns
df16.drop(columns=["Rape-Revised", "Rape-Legacy"])
```

```python
# Reorganize Columns
df16 = df16[["State", "City", "Year", "Population", "Violent Crime", "Murder", "Rape", "Robbery", "Assault", "Property", "Burglary",
             "Larceny", "MV Theft", "Arson"]]
df16
```

```python
# Start on 2018 Data
df18 = pd.DataFrame(crime18)
df18
```

```python
# Rename Columns
df18.columns = ["State", "City", "Year", "Population", "Violent Crime", "Murder", "Rape", "Robbery", "Assault", "Property",
                "Burglary", "Larceny", "MV Theft", "Arson"]
```

```python
# Remove Commas from All Values
df18 = df18.replace(',','', regex=True)
df18
```

```python
# Specify Columns Requiring Data Type Change
cols2 = ["Murder", "Arson"]
```

```python
# Change Specified Columns
df18[cols2] = df18[cols2].astype(float)
pd.options.display.float_format = '{:.0f}'.format
```

```python
# Preview DataFrame
df18
```

```python
# Merge DataFrames
df15_18 = pd.concat([df16, df18])
df15_18
```

```python
# Reorganize Data
crime_merged = df15_18.copy()
chrono_years = crime_merged.sort_values(by=["Year", "State"])
chrono_years = chrono_years.reset_index()
chrono_years
```

```python
# Delete Column Containing Original Index Numbers
del chrono_years["index"]
chrono_years
```

```python
# Create CSV for Analysis
chrono_years.to_csv("Cleaned_Crime_Data.csv")
```
